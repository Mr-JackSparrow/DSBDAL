```python
import numpy as np
import pandas as pd
from sklearn import linear_model
# from sklearn.metrics import 
from sklearn.model_selection import train_test_split
```

```python
df['CRIM'].fillna(method='bfill', inplace=True)
df['ZN'].fillna(method='bfill', inplace=True)
df['INDUS'].fillna(method='bfill', inplace=True)
df['CHAS'].fillna(method='bfill', inplace=True)
df['LSTAT'].fillna(method='bfill', inplace=True)
```

```python
df.isnull().sum()
```

```python
mean = df['AGE'].mean()
df['AGE'].fillna(mean, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['AGE'] = df['AGE'].astype('int64')
```

```python
df
```

```python
x = df.drop(['MEDV'], axis=1)
y = df['MEDV']
```

```python
x.shape
```

```python
x.isnull().sum()
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=44)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
m = linear_model.LinearRegression()
m.fit(x_train, y_train)
```

```python
predicted = m.predict(x_test)
```

```python
predicted
```

```python
y_test
```

```python
MSE = round(np.mean((predicted-y_test)**2), 2)
MSE
```

```python
from sklearn.metrics import mean_squared_error
print(round(mean_squared_error(y_test, predicted)), 2)
```

```python
train_score = m.score(x_train,y_train)
print("R-Squared on Training data:", train_score) # how well the model fix to this dataset
```
